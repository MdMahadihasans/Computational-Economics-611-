```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the plot_data function
def plot_data(df, names, title='Data Plot', legend=True):
    """
    Plots 2 to 3 different variables from the DataFrame.
    
    Parameters:
    - df: DataFrame containing the data.
    - names: List of column names to plot.
    - title: Title of the plot.
    - legend: Boolean to indicate if legend should be included.
    """
    plt.figure(figsize=(10, 6))  
    for name in names:
        plt.plot(df[name], label=name)  

    plt.title(title)  
    plt.xlabel('Index')  
    plt.ylabel('Value')  

    if legend:
        plt.legend() 

    plt.show()  


df = pd.read_csv('D:/large_dataset.csv')

# Call the plot_data function with the DataFrame and the columns you want to plot
plot_data(df, ['Value1', 'Value2', 'Value3'], title='Sample Plot')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_aggregated_bar_graph(df, names, aggregation='mean', title='Aggregated Bar Graph'):
 
    aggregated_data = df[names].agg(aggregation)
    aggregated_data.plot(kind='bar', figsize=(12, 8))

    plt.title(title)  
    plt.xlabel('Variable')  
    plt.ylabel(f'Aggregated Value ({aggregation})')  
    plt.xticks(rotation=0)  
    plt.legend(title='Variables')

    plt.show()

# Load the DataFrame from a CSV file
df = pd.read_csv('D:/large_dataset.csv')

# Call the plot_aggregated_bar_graph function with the DataFrame and the columns you want to plot
plot_aggregated_bar_graph(df, ['Value1', 'Value2', 'Value3'], aggregation='mean', title='Sample Aggregated Bar Graph')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_aggregated_bar_graph(df, names, aggregation='mean', title='Aggregated Bar Graph', colors=None):
  
    # Set the style
    plt.style.use('ggplot')

    # Aggregate the data
    aggregated_data = df[names].agg(aggregation)

    # Plot configuration
    ax = aggregated_data.plot(kind='bar', figsize=(12, 8), color=colors, width=0.8)
    ax.set_title(title, fontsize=20, fontweight='bold', color='navy')
    ax.set_xlabel('Variable', fontsize=15, fontweight='bold', color='darkred')
    ax.set_ylabel(f'Aggregated Value ({aggregation})', fontsize=15, fontweight='bold', color='darkred')
    ax.tick_params(axis='x', labelsize=12, labelrotation=0, colors='darkgreen')
    ax.tick_params(axis='y', labelsize=12, colors='darkgreen')


    ax.legend(title='Variables', fontsize=12, title_fontsize='13', loc='upper right')


    ax.grid('on', which='major', axis='y', linestyle='--', linewidth=0.5)

    plt.show()  # Displays the plot


df = pd.read_csv('D:/large_dataset.csv')
colors = ['skyblue', 'lightgreen', 'salmon']


plot_aggregated_bar_graph(df, ['Value1', 'Value2', 'Value3'], aggregation='mean', title='Sample Aggregated Bar Graph', colors=colors)
```
