## Data Frame from Dictionary

```python
import pandas as pd

# Define a dictionary where each key-value pair will become a column in the DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 35, 40, 45],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
}

# Create a DataFrame from the dictionary
df = pd.DataFrame(data)
print(df)
```

```python


# Initialize the dictionary with keys as column names and empty lists as values
macro_dict = {
    "GDP": [],
    "Money": [],
    "Real GDP": [],
    "Price Level": []
}
macro_dict
```

```python
for key in macro_dict:
    macro_dict[key] = [round(np.random.random() * 10000, 2) for i in range(1990, 2019)]

# Print the populated dictionary
print(macro_dict)
```

```python
pd.DataFrame(macro_dict)
```

```python
macro_df=pd.DataFrame(macro_dict)
macro_df.tail()
```

```python
macro_df["Velocity"] = macro_df["GDP"] / macro_df["Money"]
print(macro_df)
```

```python
macro_df.loc[4:17,["Real GDP", "Money"]]
```

```python
macro_df.loc[[4,5,8],["Real GDP", "Money"]]
```

```python
macro_df.loc[:,["Real GDP"]]
```

```python
income_data = income_data.rename(columns={'Unnamed: 0': 'Household Income'})

# Displaying the first few rows of the DataFrame to check the new column names
income_data.head()
```

```python
# Selecting the 'Number' column from the DataFrame
income = income_data['Number'][1:-1].div(1000)

# Displaying the selected 'Number' column
income
```

```python
# Dividing the 'Number' column by 1000, starting from the second row
income = income_data['Number'][1:].div(1000)

# Setting the first value of the 'income' Series to 0
income.iloc[0] = 0

# Displaying the first few entries in the 'income' Series to verify changes
income.head()
```

```python
# Creating a list of income levels, presumably for binning or labeling purposes
income_levels = list(range(0, 201, 5))

# Appending an additional level to the 'income_levels' list
income_levels.append(250)

# Converting each number in 'income_levels' to a string and appending 'k' to denote 'thousands'
income_levels = [str(value) + "k" for value in income_levels]

# Printing the 'income_levels' list
print(income_levels)
```
